```python
import pandas as pd
import random
food_calories = {
    "grains": {
        "Rice": 200,
        "Noodles": 220,
    },
    "proteins": {
        "Egg": 70,
        "Tofu": 80,
        "Beans": 130
    },
}
```

```python
data = []
for category, items in food_calories.items():
    for food, calories in items.items():
        data.append([category, food, calories])

df = pd.DataFrame(data, columns=["Category", "Food", "Calories"])
print(df)
```
